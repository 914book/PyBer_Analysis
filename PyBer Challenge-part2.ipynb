```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
pyber_data_df = pyber_data_df.rename(columns={'city': 'City', 'date': 'Date', 'fare': 'Fare', 'ride_id': 'Ride ID', 'driver_count': 'No.Drivers', 'type': 'City Type'}) 
pyber_data_df.head()
```

```python
pyber_data_df.set_index(pyber_data_df['Date'], drop=False, append=False, inplace=False, verify_integrity=False).drop('Date', 1)
```

```python
pyber_data_new_df = pyber_data_df[['Date', 'City Type', 'Fare']].copy()
pyber_data_new_df
```

```python
datetime_df = pyber_data_new_df.set_index(pyber_data_new_df['Date'], drop=False, append=False, inplace=False, verify_integrity=False).drop('Date', 1)
```

```python
datetime_df


```

```python
# Total fares for each city type.
sum_fares_by_type = pyber_data_new_df.groupby(['City Type', 'Date']).sum()["Fare"]
sum_fares_by_type
```

```python
fares_by_citydate_df = pd.DataFrame(sum_fares_by_type)
fares_by_citydate_df
```

```python
fares_by_citydate_df.reset_index()
```

```python
# Create the Urban/suburban/rural city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] =="Urban"]
suburban_cities_df =pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
urban_cities_df.head()
rural_cities_df.head()
suburban_cities_df.head()

```

```python
city_type = city_data_df["type"].unique()
city_type
```

```python
# Total Rides per city type

sum_rides_by_type=pyber_data_df.groupby(["type"]).count()["ride_id"]
sum_rides_by_type
```

```python
# Total Drivers per city type

sum_drivers_by_type=city_data_df.groupby(["type"]).sum()["driver_count"]
sum_drivers_by_type
```

```python
# Total fares for each city type.
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
# Average Fare per Ride by city type

average_fare_by_type = sum_fares_by_type/sum_rides_by_type
average_fare_by_type.map("{:.2f}".format)
```

```python
# Average Fare per Driver by city type

average_fare_Driver_by_type = sum_fares_by_type/sum_drivers_by_type
average_fare_Driver_by_type.map("{:.2f}".format)
```

```python
# Adding a list of values with keys to create a new DataFrame.
final_summary_df = pd.DataFrame(
          [{"Total Rides": sum_rides_by_type,
          "Total Drivers": sum_drivers_by_type,
          "Total Fares": sum_fares_by_type,
          "Average Fare per Ride": average_fare_by_type,
          "Average Fare per Driver": average_fare_Driver_by_type}])
final_summary_df
```

```python
# Format the columns
final_summary_df["Total Rides"] = sum_rides_by_type
```
